## Week 2 : Frequentist Inference : Plotting likelihood

Plotting the likelihood and log-likelihood function for a Binomial Distribution using Python libraries instead of R or Excel.

Note that the likelihood function omits a constant value <sup>n</sup>C<sub>y</sub> because it is harder to compute and does not change the shape of the curve or position of the MLE (Maximum Likelihood Estimate).

```python
import matplotlib.pyplot as plt
import numpy as np
import pymc3 as pm
%matplotlib inline
```

```python
def likelihood(n, y, theta):
    """ we assume a Binomial distribution """
    return (theta ** y) * (1 - theta) ** (n - y)

theta = np.arange(0.01, 0.99, 0.01)
likelihood(400, 72, theta)
```

```python
like = likelihood(400, 72, theta)
plt.plot(theta, like)
plt.xlabel("theta")
plt.ylabel("likelihood")
theta_hat = theta[np.argmax(like)]
plt.axvline(x=theta_hat, color="r")
```

```python
def log_likelihood(n, y, theta):
    return y * np.log(theta) + (n - y) * np.log(1 - theta)

log_like = log_likelihood(400, 72, theta)
plt.plot(theta, log_like)
plt.xlabel("theta")
plt.ylabel("log likelihood")
theta_hat = theta[np.argmax(log_like)]
plt.axvline(x=theta_hat, color="r")
```
