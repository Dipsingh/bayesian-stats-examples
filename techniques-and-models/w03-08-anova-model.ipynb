# ANOVA Model using PyMC3

ANalysis Of VAriance (ANOVA) methods are used when we have categorical explanatory variables in our linear model. It compares the variability of the response variable across groups to the variability within group. If the value is large, we can conclude there is a grouping effect, i.e., there is an effect of the categorical variable.

We will use the [PlantGrowth dataset](http://vincentarelbundock.github.io/Rdatasets/doc/datasets/PlantGrowth.html) from the RDatasets repository.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pymc3 as pm
import scipy.stats as stats
import statsmodels.api as sm
%matplotlib inline
```

## Visualize Data

Based on visualization below, we have the following findings:

* Group `trt2` has the highest mean yield.

```python
categories = plantgrowth_df["group"].unique()
data = [plantgrowth_df[plantgrowth_df["group"] == cat]["weight"].values for cat in categories]
plt.boxplot(data, labels=categories)
_ = plt.show()
```

## Baseline Linear Model

We cannot use the statsmodels OLS linear regression module with the group variables as they are, so we need to convert them to dummy variables.

Findings from the model:
* Mean of `ctrl` group is `5.03`, Mean of `trt1` is `4.66 (5.03 - 0.37)` and mean of `trt2` is `5.5 (5.03 + 0.49)`

```python
plantgrowth_df = pd.concat((plantgrowth_df, 
                            pd.get_dummies(plantgrowth_df["group"], prefix="group", drop_first=True)), 
                           axis=1)
plantgrowth_df.head()
```

```python
x = plantgrowth_df[["group_trt1", "group_trt2"]].values
x0 = np.ones((len(x), 1))
X = np.hstack((x0, x))

y = plantgrowth_df["weight"].values

x.shape, x0.shape, X.shape, y.shape
```

```python
base_model = sm.OLS(y, X).fit()
base_model.summary()
```

## One-way ANOVA

R has a way to generate an ANOVA table on the linear model. The `statsmodels` package also seems to have a way to do this but I was not able to get this to work. 

The `scipy.stats.f_oneway` function tests if the null hypothesis that multiple groups have the same population mean, and the values of `statistic` and `pvalue` correspond to the `F-value` and `Pr(>F)` values returned by the R function respectively. Here the `F-value` is the quotient of the variance across groups and the variance within groups and high-values are good, and the `Pr(>F)` is the p-value of 0.016 which is marginally significant.

```python
stats.f_oneway(data[0], data[1], data[2])
```

## MCMC Model 1: Cell Means Model

In this model, each group has its own mean. Since we don't want to make any assumptions about what the mean should be, we will supply a broad uninformative prior for each mean (`mu_0`).

Findings:

* Convergence -- the model has converged.
  * Traceplots look good
  * Gelman Rubin diagnostic looks good
  * Autocorrelation plots look good
  * Effective sample sizes are large
* Posterior means estimates
  * Output of `pm.summary` matches our estimates for mean from our OLS model.
* Residuals
  * No pattern seen in plot for residuals
  * In prediction vs residual plot, residual variance for `ctrl` group is much higher than residual variance for `trt2` group. May be more appropriate to have additional variance parameters for each group.

```python
cat_dict = {"ctrl": 0, "trt1": 1, "trt2": 2}
group_idx = [cat_dict[x] for x in plantgrowth_df["group"].values]
```

```python
init_params = {
    "alpha_0": 2.5,
    "beta_0": 2.5,
    "means_mu_0": 0, 
    "means_sig_0": 1e6,
    "mu_0": stats.norm.rvs(loc=0, scale=100, size=3),
    "sig_0": stats.gamma.rvs(1, scale=1, size=1)[0]
}
n_tune = 1000
n_iter = 5000

cell_mean_model = pm.Model()
with cell_mean_model:
    prec = pm.InverseGamma("prec", alpha=init_params["alpha_0"], beta=init_params["beta_0"])
    means = pm.Normal("means", mu=init_params["means_mu_0"], sigma=init_params["means_sig_0"], shape=3)

    sd = np.sqrt(1 / prec)
    y_obs = pm.Normal("y_obs", mu=means[group_idx], sigma=sd, observed=y)
    
    trace_cm = pm.sample(n_iter, tune=n_tune)
```

```python
_ = pm.traceplot(trace_cm, combined=True)
```

```python
pm.gelman_rubin(trace_cm)
```

```python
_ = pm.autocorrplot(trace_cm)
```

```python
pm.effective_n(trace_cm)
```

```python
pm.summary(trace_cm)
```

```python
group_coeffs = np.mean(trace_cm.get_values("means"), axis=0)
preds = [group_coeffs[g] for g in group_idx]
resids = y - preds
```

```python
plt.scatter(np.arange(len(resids)), resids)
plt.xlabel("index")
plt.ylabel("residuals")
_ = plt.show()
```

```python
plt.scatter(preds, resids)
plt.xlabel("predictions")
plt.ylabel("residuals")
_ = plt.show()
```

### Highest Posterior Densities

Default credible interval is 94%, can be changed appropriately using the `credible_interval` attribute.

```python
means = trace_cm.get_values("means")
pm.hpd(means)
```

### Query Model: Does any of the treatments increase mean yield of plants?

* It is clear from the posterior summaries that `trt1` does not increase the mean yield compared to the `ctrl`, but `trt2` might.
* Since Bayesian models give us a distribution, it is easy to compute posterior probabilities of such hypotheses.

Probability of whether `trt2` increases the mean yield of plants compared to `ctrl`:

```python
np.mean(means[:, 2] > means[:, 0])
```

### Query Model: Probability of trt2 increasing mean yield by at least 10%

Assume `trt2` is expensive and in order to make the increased yield worthwhile, its yield must be at least 10% greater than the mean yield from `ctrl`.

This is computed as follows:

```python
np.mean(means[:, 2] > 1.1 * means[:, 0])
```

### Warning

The above practice of querying MCMC models with arbitrary hypotheses is generally not encouraged and can lead to spurious results by the law of averages. It is recommended to build your hypothesis first, build a model, run it and answer these hypotheses. Results tend to be more reliable with this approach.
