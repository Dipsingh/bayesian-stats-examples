# Poisson Regression Quiz

Dataset: [badhealth dataset](http://vincentarelbundock.github.io/Rdatasets/doc/COUNT/badhealth.html).

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pymc3 as pm
import scipy.stats as stats
import statsmodels.api as sm

from scipy.special import expit

%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
try:
    badhealth_df = pd.read_csv("badhealth.csv", index=False)
except:
    badhealth_df = pd.read_csv("http://vincentarelbundock.github.io/Rdatasets/csv/COUNT/badhealth.csv")
    badhealth_df.to_csv("badhealth.csv")
    
badhealth_df.head()
```

```python
len(badhealth_df), len(badhealth_df.dropna())
```

```python
model_1 = pm.Model()
with model_1:
    coeffs = pm.Normal("coeffs", mu=0, sigma=10, shape=4)

    log_preds = (coeffs[0] + 
                 coeffs[1] * badhealth_df["badh"].values + 
                 coeffs[2] * badhealth_df["age"].values + 
                 coeffs[3] * (badhealth_df["badh"] * badhealth_df["age"]).values)
    y_obs = pm.NegativeBinomial("y_obs", mu=pm.math.exp(log_preds), alpha=2.5, 
                                observed=badhealth_df["numvisit"].values)
    
    trace_1 = pm.sample(5000, tune=1000)
```

```python
_ = pm.traceplot(trace_1, combined=True, compact=False)
```

```python
pm.summary(trace_1)
```

```python
model_2 = pm.Model()
with model_2:
    coeffs = pm.Normal("coeffs", mu=0, sigma=10, shape=4)

    log_preds = (coeffs[0] + 
                 coeffs[1] * badhealth_df["badh"].values + 
                 coeffs[2] * badhealth_df["age"].values)
    
    y_obs = pm.NegativeBinomial("y_obs", mu=pm.math.exp(log_preds), alpha=2.5, 
                                observed=badhealth_df["numvisit"].values)
    
    trace_2 = pm.sample(5000, tune=1000)
```

```python
_ = pm.traceplot(trace_2, combined=True, compact=False)
```

```python
pm.summary(trace_2)
```

```python
compare_dict = {
    "model_1": trace_1,
    "model_2": trace_2
}
pm.compare(compare_dict)
```

```python
callers_df = pd.read_csv("callers.csv")
callers_df.head()
```

```python
x_age = callers_df["age"].values
x_isgroup2 = callers_df["isgroup2"].values
y = (callers_df["calls"] * callers_df["days_active"]).values
```

```python
model_3 = pm.Model()
with model_3:
    coeffs = pm.Normal("coeffs", mu=0, sigma=10, shape=3)
    log_preds = (coeffs[0] + 
                 coeffs[1] * x_age + 
                 coeffs[2] * x_isgroup2)
    y_obs = pm.NegativeBinomial("y_obs", mu=pm.math.exp(log_preds), alpha=2.5, observed=y)
    trace_3 = pm.sample(5000, tune=1000)
```

```python
pm.summary(trace_3)
```

```python
coeffs_b2 = trace_3.get_values("coeffs")[:, 2]
np.mean(coeffs_b2 > 0)
```
