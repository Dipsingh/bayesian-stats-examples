# Multiple Factor ANOVA

The course document uses the [Warpbreaks](http://vincentarelbundock.github.io/Rdatasets/doc/datasets/warpbreaks.html) dataset for this example.

Explanatory variables are `wool` with two values `A` and `B`, and `tension`, with 3 values `Low`, `Medium`, and `High`. Response varibale is the number of breaks.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pymc3 as pm
import scipy.stats as stats
import statsmodels.api as sm

from scipy.special import expit

%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore")
```

## Load Data

```python
try:
    warpbreaks_df = pd.read_csv("warpbreaks.csv", index=False)
except:
    warpbreaks_df = pd.read_csv("http://vincentarelbundock.github.io/Rdatasets/csv/datasets/warpbreaks.csv")
    warpbreaks_df.to_csv("warpbreaks.csv")
    
warpbreaks_df.head()
```

## Visualize Data

* Tension is categorical with 3 levels Low (L), Medium (M), and High (H).
* Wool is categorical with 2 levels A and B.
* Variance between groups grouped by (wool + tension) is less if we consider log(breaks).

```python
pd.crosstab(warpbreaks_df["wool"], warpbreaks_df["tension"])
```

```python
data, labels = [], []
for tension in ["L", "M", "H"]:
    for wool in ["A", "B"]:
        data.append(warpbreaks_df[(warpbreaks_df["tension"] == tension) & 
                                  (warpbreaks_df["wool"] == wool)]["breaks"].values)
        labels.append("".join([wool, tension]))

plt.boxplot(data, labels=labels)
plt.ylabel("breaks")
_ = plt.show()
```

```python
log_data = []
for d in data:
    logd = np.log(np.array(d))
    log_data.append(logd)
    
plt.boxplot(log_data, labels=labels)
plt.ylabel("log(breaks)")
_ = plt.show()
```

## One Way Model

Using `tension` as the explanatory variable only.

```python
warpbreaks1_df = pd.concat((warpbreaks_df,
                            pd.get_dummies(warpbreaks_df["tension"], prefix="tension", drop_first=True)),
                           axis=1)
warpbreaks1_df.head()
```

```python
Xdata = warpbreaks1_df.values[:, 4:]
Xone = np.ones(Xdata.shape[0]).reshape(-1, 1)
X = np.hstack((Xone, Xdata)).astype(np.float32)

y = np.log(warpbreaks1_df["breaks"].values)

Xdata.shape, Xone.shape, X.shape, y.shape
```

```python
model_1 = pm.Model()
with model_1:
    coeffs = pm.Normal("coeffs", mu=0, sigma=1e6, shape=3)
    prec = pm.InverseGamma("prec", alpha=2.5, beta=0.4)

    sd = np.sqrt(prec)
    preds = (coeffs[0] * X[:, 0] +
             coeffs[1] * X[:, 1] + 
             coeffs[2] * X[:, 2])
    
    y_obs = pm.Normal("y_obs", mu=preds, sigma=sd, observed=y)
    
    trace_1 = pm.sample(5000, tune=1000)
```

```python
_ = pm.traceplot(trace_1, combined=True, compact=False)
```

```python
pm.summary(trace_1)
```

## Two-way Additive Model

We will first fit the additive model which treats the two factors `tension` and `wool` separately with no interaction. There are 6 possible values for the 2 categorical variables taken together.

**NOTE:** Model shown here is slightly different from the one in the course, it uses both levels of `wool` (A and B). This is because the model with 1 level of `wool` was not converging properly, possibly because it tries to describe the base state for two variables `tension` and `wool` in the intercept term. In this model, the intercept term contains only the base state for `tension` (H).

```python
warpbreaks2_df = pd.concat((warpbreaks_df,
                            pd.get_dummies(warpbreaks_df["tension"], prefix="tension", drop_first=True),
                            pd.get_dummies(warpbreaks_df["wool"], prefix="wool")),
                           axis=1)
warpbreaks2_df.head()
```

```python
Xdata = warpbreaks2_df.values[:, -4:]
Xones = np.ones(Xdata.shape[0]).reshape(-1, 1)
X = np.hstack((Xones, Xdata)).astype(np.float32)

y = np.log(warpbreaks2_df["breaks"].values)

Xdata.shape, Xones.shape, X.shape, y.shape
```

```python
model_2 = pm.Model()
with model_2:
    coeffs = pm.Normal("coeffs", mu=0, sigma=1e3, shape=5)
    prec = pm.InverseGamma("prec", alpha=2.5, beta=0.4)
    
    sd = np.sqrt(prec)
    pred = (coeffs[0] * X[:, 0] +
            coeffs[1] * X[:, 1] + 
            coeffs[0] * X[:, 2] +
            coeffs[1] * X[:, 3] + 
            coeffs[2] * X[:, 4])
    
    y_obs = pm.Normal("y_obs", mu=pred, sigma=sd, observed=y)

    trace_2 = pm.sample(5000, tune=1000)
```

```python
_ = pm.traceplot(trace_2, combined=True, compact=False)
```

```python
pm.summary(trace_2)
```

## Cell-Means Model

In the additive model, the effects of `wool_A (coeffs[3])` and `wool_B (coeffs[4])` are both negative. with `wool_A` being more negative than `wool_B`. This does not seem to be borne out by the box plot, since for `tension_M`, `wool_A` seems to have lower `log(breaks)` than `wool_B`, but it is the opposite for the other `tension` classes. This leads us to suspect that there might be some interaction effects. The cell means model tries to account for interaction features, by giving each of the 6 `wool*tension` subgroups its own prior.

```python
warpbreaks3_df = pd.DataFrame.copy(warpbreaks_df, deep=True)
warpbreaks3_df.head()
```

```python
for wool in ["A", "B"]:
    for tension in ["L", "M", "H"]:
        colname = "wt_{:s}{:s}".format(wool, tension)
        warpbreaks3_df[colname] = ((warpbreaks3_df["wool"] == wool) & 
                                   (warpbreaks3_df["tension"] == tension)).astype(int)
warpbreaks3_df.head()
```

```python
X = warpbreaks3_df.values[:, -6:].astype(np.float32)
y = np.log(warpbreaks3_df["breaks"].values)

X.shape, y.shape
```

```python
model_3 = pm.Model()
with model_3:
    prec = pm.InverseGamma("prec", alpha=1.5, beta=0.67)
    coeffs = pm.Normal("coeffs", mu=0, sigma=1e3, shape=6)

    sd = np.sqrt(prec)
    preds = (coeffs[0] * X[:, 0] +
             coeffs[1] * X[:, 1] + 
             coeffs[2] * X[:, 2] + 
             coeffs[3] * X[:, 3] + 
             coeffs[4] * X[:, 4] + 
             coeffs[5] * X[:, 5])
    
    y_obs = pm.Normal("y_obs", mu=preds, sigma=sd, observed=y)

    trace_3 = pm.sample(5000, tune=1000)
```

```python
_ = pm.traceplot(trace_3, combined=True, compact=False)
```

```python
pm.summary(trace_3)
```

### Find best treatment group

We look at the trace of the posteriors, and for each generated sample, we compute the one with lowest value (the lowest `log(break)`. Counting these minimum values will give us the "best treatment group".

Here out best treatment group looks like `wool_B` and `tension_H`, the same as that found in the course document.

```python
coeffs = trace_3.get_values("coeffs", combine=True)
unique, counts = np.unique(np.argmin(coeffs, axis=1), return_counts=True)
[(u, c) for u, c in zip(unique, counts)]
```

## Compare Models

Findings:
* In terms of WAIC, `model_1` appears to be the best model, but only marginally compared to `model_2`. The course document finds that `model_2` is the best in terms of DIC.

```python
compare_dict = {
    "model_1": trace_1,
    "model_2": trace_2,
    "model_3": trace_3
}
comp = pm.compare(compare_dict)
comp
```

```python
_ = pm.compareplot(comp)
```
