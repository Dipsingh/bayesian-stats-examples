# Linear Regression (single variable) - PyMC3

We will recreate the Linear Regression model using the same data as [the previous notebook](http://localhost:8888/notebooks/w03-01a-linreg-demo.ipynb) but using PyMC3 this time.

## Problem Definition

We want to predict the infant mortality rate given the parent's income level. We will use the [Leinhardt toy dataset](https://vincentarelbundock.github.io/Rdatasets/doc/carData/Leinhardt.html) for this.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pymc3 as pm
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore")
```

## Load Data

```python
try:
    leinhardt_df = pd.read_csv("Leinhardt.csv")
except:
    leinhardt_df = pd.read_csv("https://vincentarelbundock.github.io/Rdatasets/csv/carData/Leinhardt.csv")
    leinhardt_df.to_csv("Leinhardt.csv")
    
leinhardt_df.head()
```

## Transform and Clean Data

We already know from previous notebook that the relation between `income` and `infant` (mortality) is kind of linear on the log scale, and that there are 4 records that need to be removed. We do that below, and extract the X and Y variables into numpy arrays.

```python
leinhardt_df["log_income"] = np.log(leinhardt_df["income"])
leinhardt_df["log_infant"] = np.log(leinhardt_df["infant"])

leinhardt_df.dropna(subset=["log_income", "log_infant"], inplace=True)
```

```python
x = leinhardt_df["log_income"].values
y = leinhardt_df["log_infant"].values
x.shape, y.shape
```

## Define Model

Our objective is to fit a linear regression model of the form (in this case N=1).

$$ y = \beta_0 + \sum_{j=0}^{N} \beta_j * x_j $$

where we assume that y comes from a Normal distribution parameterized by mean $\mu$ and standard deviation $\sigma$. We assume also that $\sigma$ comes from an Inverse Gamma distribution with prior sample size of 5 and prior guess for variance of 10, giving us shape of `5/2 (2.5)` and scale of `2/(5*10) = 0.04`. Finally, since we don't know what the coefficients are, we will assume they come from a Normal distribution with mean `0` and a large variance `1e6` to simulate a non-informative prior.

$$ y \sim Normal(\mu, \sigma^2) $$
$$ \sigma \sim InvGamma(2.5, 0.04) $$
$$ \beta \sim Normal(0, 1e6) $$

Finally, because y comes from a Normal distribution with mean $\mu$, we can write $\mu$ in terms of the coefficient values $\beta$ as follows.

$$ \mu = \beta_0 + \beta_1 * x $$


## Sample from Model

```python
init_params = {
    "mu_0": 0,
    "sigma_0": 1e6,
    "alpha_0": 2.5,
    "beta_0": 0.04
}

n_tune = 1000
n_iter = 5000
n_chains = 3
```

```python
with pm.Model() as linreg_model:
    beta = pm.Normal("beta", mu=init_params["mu_0"], sigma=init_params["sigma_0"], shape=2)
    sigma2 = pm.InverseGamma("sigma2", alpha=init_params["alpha_0"], beta=init_params["beta_0"])

    sigma = np.sqrt(sigma2)
    mu = beta[0] + beta[1] * x
    
    y_obs = pm.Normal("y_obs", mu=mu, sigma=sigma, observed=y)
    
    trace = pm.sample(n_iter, tune=n_tune, chains=n_chains)
```

```python
_ = pm.traceplot(trace, combined=True)
```

```python
pm.summary(trace)
```
