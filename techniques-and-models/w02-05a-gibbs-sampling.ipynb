# Gibbs Sampling with Normal Likelihood

This notebook attempts to solve a variant of the problem solved manually in [w02-04b-mcmc-demo-continuous.ipynb](http://localhost:8888/notebooks/w02-04b-mcmc-demo-continuous.ipynb), and with PyMC3 in [w03-04c-jags-example.ipynb](http://localhost:8888/notebooks/w02-04c-jags-example.ipynb). The difference this time is that we allow both the mean $\mu$ and variance $\sigma^2$ to be random variables.

## Problem Definition

Data is for personnel change from last year to this year for 10 companies. Model is defined as follows. The distributions for $\mu$ and $\sigma^2$ have been chosen assuming that the other one is going to be a constant, which is true in case of Gibbs sampling, where we sample one parameter and hold all the others constant at each iteration. Here $\mu$ and $\sigma^2$ are known as conditionally conjugate priors.

$$ y_i | \mu, \sigma^2 \overset{iid}{\sim} Normal(\mu, \sigma^2) $$
$$ \mu \sim Normal(\mu_0, \sigma_0^2) $$
$$ \sigma^2 \sim InvGamma(\nu_0, \beta_0) $$

where y<sub>i</sub> represents personnel change for company i.

For the manual solution, we analytically find that the distributions for $\mu$ given $\sigma^2$ and y, and for $\sigma^2$ given $\mu$ and y, are as follows. We won't need this for the PyMC3 solution.

$$ P(\mu|\sigma^2,y) \sim Normal \left( \mu \bigg| \frac{n\overline{y}/\sigma^2 + \mu_0/\sigma_0^2}{n/\sigma^2 + 1/\sigma_0^2}, \frac{1}{n/\sigma^2 + 1/\sigma_0^2} \right ) $$
$$ P(\sigma^2|\mu, y) \sim InvGamma \left( \sigma^2 \bigg | \nu_0 + \frac{n}{2}, \beta_0 + \frac{\sum_{i=1}^n (y_i - \mu)^2}{2}  \right ) $$

### Model Specification

```python
def update_mu(n, ybar, sig2, mu_0, sig2_0):
    denom = (n / sig2) + (1 / sig2_0)
    loc = (n * ybar / sig2) + (mu_0 / sig2_0)
    scale = np.sqrt(1 / denom)
    return stats.norm.rvs(loc=loc, scale=scale, size=1)[0]


def update_sig2(n, y, mu, nu_0, beta_0):
    a = nu_0 + (n/2)
    b = beta_0 + (np.sum(np.power(y - mu, 2)) / 2)
    return stats.invgamma.rvs(a, scale=b, size=1)[0]


def gibbs(y, n_iter, init_params, priors):
    ybar = np.mean(y)
    n = len(y)
    
    mu_out = np.zeros((n_iter))
    sig2_out = np.zeros((n_iter))
    
    mu_now = init_params["mu"]
    for i in range(n_iter):
        sig2_now = update_sig2(n, y, mu_now, nu_0=priors["nu_0"], beta_0=priors["beta_0"])
        mu_now = update_mu(n, ybar, sig2_now, mu_0=priors["mu_0"], sig2_0=priors["sig2_0"])
        sig2_out[i] = sig2_now
        mu_out[i] = mu_now
        
    return mu_out, sig2_out
```

### Model Setup

```python
y = np.array([1.2, 1.4, -0.5, 0.3, 0.9, 2.3, 1.0, 0.1, 1.3, 1.9])

priors = {
    "mu_0": 0,        # set to 0 based on looking at data
    "sig2_0": 1.0,    # similar to the prior in the t distribution
#     "n_0": 2,         # guess of effective sample size
#     "s2_0": 1,        # guess of effective standard deviation
    "nu_0": 1,        # n_0 / 2
    "beta_0": 1.0     # n_0 * s2_0 / 2
}
```

### Visualize Data

Plot below shows that the main probability mass for our prior is centered at 0, but our data mean is close to 1, so we expect our posterior mean to have most of its probability mass centered at some point between 0 and 1.

```python
plt.xlim((-1.0, 3.0))
# histogram of y values
plt.hist(y, density=True, bins=5, label="y")
# actual data points along x axis
plt.scatter(y, np.zeros((len(y))), marker='o', color='r')
# sample mean
plt.axvline(np.mean(y), 0, 1, color='r', linestyle='--', label="sample mean")
xs = np.arange(-1.0, 3.0, 0.1)
plt.plot(xs, stats.norm.pdf(xs, loc=priors["mu_0"], scale=np.sqrt(priors["sig2_0"])), 
         color='g', label="prior")
plt.legend(loc="best")
_ = plt.show()
```

### Run Sampling

```python
np.random.seed(53)

n_iter = 1500   # we will discard first 500 "warmup" iterations
init_params = { "mu": 0 }

mu_out, sig2_out = gibbs(y, n_iter, init_params, priors)
```

## PyMC3 Solution

The R version shown in the video has mean posterior values of 0.9051 and 0.9282 for $\mu$ and $\sigma^2$ respectively. Here we get values that are close to these values. Also the posterior value of $\mu$ is in the range we expect it to be from visualizing the means of the prior estimate of $\mu$ and the data.

```python
mu_0 = 0
sigma2_0 = 1.0
alpha_0 = 1
beta_0 = 2.5

y = np.array([1.2, 1.4, -0.5, 0.3, 0.9, 2.3, 1.0, 0.1, 1.3, 1.9])

n_iter = 1000
n_tune = 500

with pm.Model() as model:
    mu = pm.Normal("mu", mu=mu_0, sigma=sigma2_0)
    sigma2 = pm.InverseGamma("sigma2", alpha=alpha_0, beta=beta_0)
    y_obs = pm.Normal("y_obs", mu=mu, sigma=np.sqrt(sigma2), observed=y)
    
    # optimal step function assigned by PyMC3 according to distribution
    trace = pm.sample(n_iter, tune=n_tune)
```

```python
_ = pm.traceplot(trace, combined=True)
```
